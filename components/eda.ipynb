```python
import pandas as pd
```

```python
path = "../OPEX_speakers.csv"
data = pd.read_csv(path)
```

```python
data
```

```python
data.dropna(subset=['day', 'time'], inplace=True)
data = data.fillna("")
```

```python
from datetime import datetime
def convert_to_bq_datetime(date_str, time_str):
  """Concatenates date and time strings, parses into datetime, and formats for BigQuery."""
  datetime_str = str(date_str) + ' ' + str(time_str)
  dt_object = datetime.strptime(datetime_str, '%Y-%m-%d %H:%M:%S')
  return dt_object.strftime('%Y-%m-%d %H:%M:%S')

# Apply the function to create the new column
data['bq_datetime'] = data.apply(lambda row: convert_to_bq_datetime(row['day'], row['time']), axis=1)
```

```python
import json

def save_data_as_jsonl(data, filename):
    """Saves data as a JSONL file.

    Args:
        data: A list of dictionaries, where each dictionary represents a data point.
        filename: The name of the file to save the data to.
    """

    with open(filename, 'w') as f:
        for item in data:
            f.write(json.dumps(item) + '\n')
data["position"] = data.reset_index().index
data["day"] = data['day'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d').strftime('%Y-%m-%d'))
data['time'] = data["time"].apply(lambda x: datetime.strptime(str(x), '%H:%M:%S').strftime('%H:%M:%S'))
data.to_csv("../data.csv")
data = data.to_dict(orient = "records")

save_data_as_jsonl(data,"../data_json.jsonl")
```
